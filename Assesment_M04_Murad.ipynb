```python
customers = {'customer_id': ['C01', 'C02', 'C03', 'C04', 'C05', 'C06', 'C07', 'C08', 'C09', 'C10'],
            'name': ['AJG', 'PMG', 'LUV', 'ZAM', 'MJA', 'ASB', 'LKM', 'ADF','JES', 'RCG'],
            'country': ['Spain', 'France', 'Libia', 'Portugal','France', 'Libia','Spain', 'France', 'Portugal', 'Spain' ],
             'contract_type': ['C1', 'C2', 'C3', 'C2', 'C3', 'C1','C3', 'C1', 'C2', 'C1'  ],
             'age': [33, 25, 44, 27, 51, 26, 62, 34, 19, 37 ],
             'invoice1': [100, 125, 144, 117, 91, 86, 82, 94, 119, 131 ],
             'invoice2': [103, 115, 111, 118, 98, 88, 88, 93, 109, 131 ],
             'invoice3': [90, 105, 131, 110, 95, 91, 79, 99, 112, 135 ]
            }
df_MST = pd.DataFrame(customers)
df_MST
```

1)Create a new __calculated field__ on df_xyz. The calculated field will be obtained based on a formula or expression __chosen by the student__ (each student must choose his or her formula). The name of the new field will be f_xyz (where xyz are the initials of the student name)

```python
df_MST['f_MST'] = 5*df_MST['invoice1'] - df_MST['invoice2'] + df_MST['invoice3']
df_MST
```

```python

from matplotlib import pyplot as plt
df_MST['invoice3'].plot(kind='hist', bins=20, title='invoice3')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

3)__Perform an aggregation operation__ using contract_type as grouping field.Perform an aggregation operation over the f_xyz field. The student will choose the aggregation operation (count(), sum(), mean(), etc.). The name of the new calculated field (result of the aggregation) will be a_xyz
